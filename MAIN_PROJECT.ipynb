```python
import pandas as pd
from scipy import stats

# Load the first CSV file
df1 = pd.read_csv('/content/drive/MyDrive/covid-19-confirm-cases2.csv')
print(df1)
# Load the second CSV file
df2 = pd.read_csv('/content/drive/MyDrive/covid-19-death-cases1.csv')
print(df2)
# Assuming both CSV files have a column named 'value' representing the data
data1 = df1['Lat']
data2 = df2['Lat']

# Hypothesis test (Two-sample t-test)
t_stat,p_value= stats.ttest_ind(data1,data2)

print(f'Two-sample t-test:\n  t-statistic = {t_stat}\n  p-value = {p_value}')

# ANOVA test (Analysis of Variance)
f_stat, p_value_anova = stats.f_oneway(data1, data2)
print(f'ANOVA test:\n  F-statistic = {f_stat}\n  p-value = {p_value_anova}')
```
