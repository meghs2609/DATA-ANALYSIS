```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
print(tips)
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("scanner plot of total bill vs Tip")
plt.xlabel("Total bill ($)")
plt.ylabel("Tip ($)")
plt.show
```

```python
a=sns.load_dataset("anscombe")
print(a)
sns.jointplot(x="x",y="y",data=a)
plt.title("Anscombes quartet dataset")
plt.xlabel("x ($)")
plt.ylabel("y ($)")
plt.show
```

```python
a=sns.load_dataset("attention")
print(a)
sns.jointplot(x="subject",y="attention",data=a)
plt.title("Response times in a psychological experiment")
plt.xlabel("subject ($)")
plt.ylabel("attention($)")
plt.show
```
