```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
print(tips)
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("scanner plot of total bill vs Tip")
plt.xlabel("Total bill ($)")
plt.ylabel("Tip ($)")
plt.show
```

```python
iris=sns.load_dataset("iris")
print(iris)
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris)
plt.title("scanner plot sepal_length vs sepal_width")
plt.xlabel("sepal_length ($)")
plt.ylabel("sepal_width ($)")
plt.show
```

```python
iris=sns.load_dataset("iris")
correlation_matrix=iris.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")#graph in different format=heatmap
plt.title("correlation Heatmap of Iris Dataset")
plt.show()
```

```python
titanic=sns.load_dataset("titanic")
print(titanic)
g=sns.FacetGrid(titanic,col="class")
g.map(sns.histplot,"age")
plt.subplots_adjust(top=0.8)
g.fig.suptitle("Age Distribution by Passanger Class")
plt.show
```
